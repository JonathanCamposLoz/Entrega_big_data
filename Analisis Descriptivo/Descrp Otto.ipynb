```python
import pandas as pd
import matplotlib.pyplot as plot
```

```python

df =pd.read_csv('complaint_data.csv')
df
```

### tipos de variables

* Date recieved = cuantitativa continua (esto no puede llevarse como cuantitativo)
* Product = Cualitativa ordinal
* Sub-product = Cualitativa ordinal
* Issue = Cualitativa ordinal
* Sub-issue = Cualitativa ordinal
* Consumer complaint narrative = Cualitativa nominal
* Company public response = Cualitativa ordinal
* Company	= Cualitativa nominal (puede ser ordinal si se pone un numero por compañia)
* State = Cualitativa ordinal (puede ser ordinal si se pone un numero por compañia)
* ZIP cod= Cuantitativa discreta
* Tags = Cualitativa ordinal
* Consumer consent provided? = Cualitativa ordinal
* Date sent to company = cuantitativa continua
* Company response to consumer = booleana
* Timely response? = cualitativa ordinal
* Consumer disputed? = cualitativa ordinal
* Complaint ID = Cualitativa nominal (puede ser ordinal)

Dentro del análisis exploratorio se lleva a cabo un cálculo de frecuencias relativas para cada variable cualitativa, con la finalidad de determinar el comportamiento de las diferentes variables individuales.

## Tablas y graficos 

### Producto

```python
product=pd.DataFrame(df['Product'].value_counts())
product
```

```python
plot = product.plot(kind='barh',color='skyblue',title='Complaints by product',figsize=(6,6),xlabel='Complaints', ylabel='Product',grid=True,)
```
